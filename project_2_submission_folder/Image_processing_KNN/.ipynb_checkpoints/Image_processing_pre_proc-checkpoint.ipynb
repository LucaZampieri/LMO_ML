# Hough transform + Edge detection

Trying to find another method to delimitate the streets we tried with edge detection. Moreover, we tried to empower this method with another technique named "Probabilistic Hough Transform" able to detect only the straight edges. The goal  of this procedure consisted in finding the patches which contained this straight lines and characterize the number of consecutive neighbors which mantain this orientation. For instance, in an ideal scenario all the patches along a border of the streets should mantain the orientation of the edge and the length of edge ( that can be evalueted in "number of consecutive patches" ) characterize strongly this "border" patch. Altough this is true in theory, in practice we have a very chaotic scenario where the streets are interrupted by trees and other objects and so it could be very difficult to implement a robust algorithm. 

```python
import numpy as np

from skimage.transform import (hough_line, hough_line_peaks,
                               probabilistic_hough_line)
from skimage.feature import canny
from skimage import data

import matplotlib.pyplot as plt
from matplotlib import cm





# Constructing test image
image = np.zeros((100, 100))
idx = np.arange(25, 75)
image[idx[::-1], idx] = 255
image[idx, idx] = 255
image = RGB_to_grey(imgs[img_index]) 

# Classic straight-line Hough transform
h, theta, d = hough_line(image)

# Generating figure 1
fig, axes = plt.subplots(1, 3, figsize=(15, 6),
                         subplot_kw={'adjustable': 'box-forced'})
ax = axes.ravel()

ax[0].imshow(image, cmap=cm.gray)
ax[0].set_title('Input image')
ax[0].set_axis_off()

ax[1].imshow(np.log(1 + h),
             extent=[np.rad2deg(theta[-1]), np.rad2deg(theta[0]), d[-1], d[0]],
             cmap=cm.gray, aspect=1/1.5)
ax[1].set_title('Hough transform')
ax[1].set_xlabel('Angles (degrees)')
ax[1].set_ylabel('Distance (pixels)')
ax[1].axis('image')

ax[2].imshow(image, cmap=cm.gray)
for _, angle, dist in zip(*hough_line_peaks(h, theta, d)):
    y0 = (dist - 0 * np.cos(angle)) / np.sin(angle)
    y1 = (dist - image.shape[1] * np.cos(angle)) / np.sin(angle)
    ax[2].plot((0, image.shape[1]), (y0, y1), '-r')
ax[2].set_xlim((0, image.shape[1]))
ax[2].set_ylim((image.shape[0], 0))
ax[2].set_axis_off()
ax[2].set_title('Detected lines')

plt.tight_layout()
plt.show()

# Line finding using the Probabilistic Hough Transform
#image = data.camera()
image = RGB_to_grey(imgs[img_index]) 
edges = canny(image, 1, 2, 1000)
edges = canny(image, sigma=1.5, low_threshold=0.3, high_threshold=0.5, mask=None, use_quantiles=False)
#edges = feature.canny(image,sigma=1.5)

lines = probabilistic_hough_line(edges, threshold=20, line_length=4,
                                 line_gap=3)

# Generating figure 2
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharex=True, sharey=True)
ax = axes.ravel()

ax[0].imshow(image, cmap=cm.gray)
ax[0].set_title('Input image')

edges 
ax[1].imshow(edges, cmap=cm.gray)
ax[1].set_title('Canny edges')

ax[2].imshow(edges * 0)
for line in lines:
    p0, p1 = line
    ax[2].plot((p0[0], p1[0]), (p0[1], p1[1]))
ax[2].set_xlim((0, image.shape[1]))
ax[2].set_ylim((image.shape[0], 0))
ax[2].set_title('Probabilistic Hough')

for a in ax:
    a.set_axis_off()
    a.set_adjustable('box-forced')

plt.tight_layout()
plt.show()
```

# Template matching

Another possible solution to discern streets and background can derive 
from object recognition.
For instance object as cars are setted only on the streets and finding them means
finding a number of neighbours patch which have an high probability of being
streets (for example a gaussian 2D distribution). Similar for the trees. Also in
this case the limitation of the image processing methods compromise this procedure.
For instance the specificity of template matching made possible to detect a only 
some objects similar to the template (for example only a carof the same brand or
with similar features). The implementation of such method will require the acquiring
of a huge number of templates and cases

```python
# template 
import cv2
import numpy as np
from matplotlib import pyplot as plt

img_rgb = cv2.imread('training/images/'+'satImage_001.png')
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
template = cv2.imread('template/tree_1.png',0);


w, h = template.shape[::-1]

res = cv2.matchTemplate(img_gray,template,cv2.TM_CCOEFF_NORMED)
threshold = 0.5
loc = np.where( res >= threshold)
for pt in zip(*loc[::-1]):
    cv2.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0,0,255), 2)

plt.imshow(img_rgb)
```

# Spectrum

Different from the previous tentative of feature extraction which consider the image 
as overall and try to get morphological informations on the scenario, in this case
we focus on a single patch. The spectrum give informations about the variability 
of the rgb signal inside the patch. In this case is thus possible to obtain some 
features about the patch as the mean of the spectrum (module and phase) or the
variance. Since the pixel are very small, this method does not guarantee an effective
improvement of the model. On the other hand its simplicity in the implementation
gives us the possibility of testing its impoact directly in the code.
